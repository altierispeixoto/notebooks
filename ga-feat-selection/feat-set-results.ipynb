```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
def plot_graph(dataset):
    fig, ax1 = plt.subplots(figsize=(12, 4), dpi=100, facecolor='w', edgecolor='k')
    plt.title('Feature selection ')
    color = 'tab:blue'
    ax1.set_xlabel('Generations')
    ax1.set_ylabel('NÂº Features', color=color)
    ax1.plot(dataset['generation'], dataset['n_feat'], color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:orange'
    ax2.set_ylabel('Accuracy', color=color)  # we already handled the x-label with ax1
    ax2.plot(dataset['generation'], dataset['accuracy'], color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()
```

```python
%matplotlib inline
plt.plot(dataset['best'])
plt.plot(dataset['worst'])
plt.plot(dataset['average'])
plt.show()
```

```python
columns = ['generation','population','n_feat','accuracy','worst','best','median','average','std_dev']
dataset = pd.read_csv('svc_classifier_euler_3000-statistics-file.csv',names=columns)
plot_graph(dataset)
```
