```python
param = {
    'bagging_freq': 2,          
    'bagging_fraction': 0.32,   
    'boost_from_average':'false',   
    'boost': 'gbdt',
    'feature_fraction': 0.03,   
    'learning_rate': 0.007,     
    'max_depth': -1,                
    'metric':'auc',
    'min_data_in_leaf': 60,     
    'min_sum_hessian_in_leaf': 6.0,
    'num_leaves': 3,           
    'num_threads': 8,
    'tree_learner': 'serial',   
    'objective': 'binary',   
    'verbosity': 1
}
```

```python
features = [c for c in best_train_df.columns if c not in ['ID_code', 'target']]
target = best_train_df['target']
```

```python
def run_model(folds,trdf,target,tedf,num_round,verbose_eval,early_stopping_rounds):
    
    oof = np.zeros(len(trdf))
    predictions = np.zeros(len(tedf))
    features = [c for c in trdf.columns]

    for fold_, (trn_idx, val_idx) in enumerate(folds.split(trdf.values, target.values)):
        print('Fold', fold_, 'started at', time.ctime())
        
        
        trn_data = lgb_model.Dataset(trdf.iloc[trn_idx][features], label=target.iloc[trn_idx])
        val_data = lgb_model.Dataset(trdf.iloc[val_idx][features], label=target.iloc[val_idx])
    
        clf = lgb_model.train(param, trn_data, num_round, valid_sets = [trn_data, val_data], verbose_eval=verbose_eval, early_stopping_rounds = early_stopping_rounds)
    
        oof[val_idx] = clf.predict(trdf.iloc[val_idx][features], num_iteration=clf.best_iteration)
        predictions += clf.predict(tedf[features], num_iteration=clf.best_iteration) / folds.n_splits
    
        feature_importance_df = pd.DataFrame()
        fold_importance_df = pd.DataFrame()
        fold_importance_df["Feature"] = features
        fold_importance_df["importance"] = clf.feature_importance()
        fold_importance_df["fold"] = fold_ + 1
        feature_importance_df = pd.concat([feature_importance_df, fold_importance_df], axis=0)
    
    print("CV score: {:<8.5f}".format(roc_auc_score(target, oof)))
    return clf,predictions,feature_importance_df
```

```python
num_round = 100000
verbose_eval=1000
early_stopping_rounds=3500
folds = StratifiedKFold(n_splits=10, shuffle=True, random_state=44000)
clf,predictions,feature_importance_df = run_model(folds,best_train_df[features],target,best_test_df,num_round,verbose_eval,early_stopping_rounds)
```

```python
cols = (feature_importance_df[["Feature", "importance"]]
        .groupby("Feature")
        .mean()
        .sort_values(by="importance", ascending=False)[:200].index)

best_features_sint = feature_importance_df.loc[feature_importance_df.Feature.isin(cols)]
```

```python
save_best = best_features_sint.sort_values(by="importance",ascending=False)
save_best.to_csv('best_features_plus.csv',index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(14,28))
sns.barplot(x="importance", y="Feature", data=best_features_sint.sort_values(by="importance",ascending=False))
plt.title('Features importance (averaged/folds)')
plt.tight_layout()
plt.show()
```
